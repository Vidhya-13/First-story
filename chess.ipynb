```python
# DON'T MODIFY THIS CLASS
class player_result:
    def __init__(self, name: str, points: float, resistance_points: float, sonnenborn_berger: float, black: int):
        self.name = name
        self.points = points
        self.resistance_points = resistance_points
        self.sonnenborn_berger = sonnenborn_berger
        self.black = black

    def __str__(self):
        return 'player_result(name=\'' + self.name + '\', points=' + str(self.points) + ', resistance_points=' + \
               str(self.resistance_points) + ', sonnenborn_berger=' + str(self.sonnenborn_berger) + \
               ', black=' + str(self.black) + ')'

    def __eq__(self, other):
        return self.name == other.name and self.points == other.points and \
               self.resistance_points == other.resistance_points and \
               self.sonnenborn_berger == other.sonnenborn_berger and self.black == other.black


def determine_output(input: str):
    players = {}
    rounds = input.strip().split("\n\n")
    player_names = rounds[0].strip().split("\n")
    
    # Initialize players
    for name in player_names:
        players[name] = {
            'points': 0.0,
            'resistance_points': 0.0,
            'sonnenborn_berger': 0.0,
            'black': 0,
            'opponents': []
        }

    # Process each round
    for round in rounds[1:]:
        matches = round.strip().split("\n")
        for match in matches:
            white, black, white_points, black_points = match.split()
            white_points, black_points = float(white_points), float(black_points)
            
            players[white]['points'] += white_points
            players[black]['points'] += black_points
            
            players[white]['opponents'].append(black)
            players[black]['opponents'].append(white)
            
            if black_points == 1:
                players[black]['sonnenborn_berger'] += players[white]['points']
            elif black_points == 0.5:
                players[black]['sonnenborn_berger'] += 0.5 * players[white]['points']
            
            if white_points == 1:
                players[white]['sonnenborn_berger'] += players[black]['points']
            elif white_points == 0.5:
                players[white]['sonnenborn_berger'] += 0.5 * players[black]['points']
            
            players[black]['black'] += 1
    
    # Calculate resistance points
    for name, data in players.items():
        resistance_points = sum(players[opponent]['points'] for opponent in data['opponents'])
        players[name]['resistance_points'] = resistance_points

    # Create player_result instances
    results = []
    for name, data in players.items():
        results.append(player_result(
            name,
            data['points'],
            data['resistance_points'],
            data['sonnenborn_berger'],
            data['black']
        ))

    # Sort players
    results.sort(key=lambda x: (x.points, x.resistance_points, x.sonnenborn_berger, -x.black), reverse=True)
    
    return results

# Example usage
input_str = """Anna
Bob
Charlotte
Daniel
Erik
Femke

Anna Femke 1 0
Bob Erik 0 1
Charlotte Daniel 0.5 0.5

Erik Anna 1 0
Femke Charlotte 0.5 0.5
Daniel Bob 1 0

Daniel Erik 0.5 0.5
Charlotte Anna 1 0
Bob Femke 1 0"""
```

```python
output = determine_output(input_str)
for player in output:
    print(player)
```
